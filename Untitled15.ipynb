```python
import numpy as np
import pandas as pd

def hybrid_log_transform(df, cols):
    df = df.copy()
    shifts = {}

    # Clean any existing inf/-inf from previous transforms
    df = df.replace([np.inf, -np.inf], np.nan)

    for col in cols:
        if col not in df.columns:
            continue  # skip missing cols

        # Fill NaNs temporarily to avoid errors
        col_min = df[col].min(skipna=True)
        shift = -col_min + 1 if (col_min is not None and col_min < 0) else 0
        shifts[col] = shift

        def safe_log(x):
            if pd.isna(x):
                return np.nan  # keep missing for later imputation
            if x == 0:
                return 0  # preserve exact 0 meaning
            else:
                return np.log1p(x + shift)

        df[col] = df[col].apply(safe_log)

    print("âœ… Hybrid log transform applied")
    print("Shifts used:", shifts)
    return df, shifts


# --- Usage ---
num_cols = ["Weekly_Sales", "MarkDown1", "MarkDown2", "MarkDown3",
            "MarkDown4", "MarkDown5", "CPI", "Unemployment"]

walmart_df_hybrid, train_shifts = hybrid_log_transform(walmart_df, num_cols)
walmart_test_df_hybrid, test_shifts = hybrid_log_transform(walmart_test_df, num_cols)
```

```python
def hybrid_log_transform_consistent(df, cols, shifts=None):
    df = df.copy()
    df = df.replace([np.inf, -np.inf], np.nan)

    if shifts is None:  # train case â†’ calculate shifts
        shifts = {}
        for col in cols:
            if col not in df.columns:
                continue
            col_min = df[col].min(skipna=True)
            shift = -col_min + 1 if (col_min is not None and col_min < 0) else 0
            shifts[col] = shift
    # else â†’ test case, use provided shifts

    for col in cols:
        if col not in df.columns:
            continue
        shift = shifts.get(col, 0)

        def safe_log(x):
            if pd.isna(x):
                return np.nan
            if x == 0:
                return 0
            else:
                return np.log1p(x + shift)

        df[col] = df[col].apply(safe_log)

    return df, shifts


# --- Usage ---
num_cols = ["Weekly_Sales", "MarkDown1", "MarkDown2", "MarkDown3",
            "MarkDown4", "MarkDown5", "CPI", "Unemployment"]

# Train â†’ calculate shifts
walmart_df_hybrid, train_shifts = hybrid_log_transform_consistent(walmart_df, num_cols)

# Test â†’ apply same shifts
walmart_test_df_hybrid, _ = hybrid_log_transform_consistent(walmart_test_df, num_cols, shifts=train_shifts)

print("Train shifts:", train_shifts)
```

```python
def create_features(data):
    data = data.copy()
    data["weekofyear"] = data["Date"].dt.isocalendar().week.astype(int)
    data["month"] = data["Date"].dt.month
    data["year"] = data["Date"].dt.year
    data["dayofweek"] = data["Date"].dt.dayofweek

    # Rolling means
    data["roll_mean_4"] = data.groupby(["Store","Dept"])["Weekly_Sales"].shift(1).rolling(window=4).mean()
    data["roll_mean_12"] = data.groupby(["Store","Dept"])["Weekly_Sales"].shift(1).rolling(window=12).mean()

    # Lags
    for lag in [1, 2, 3, 52]:
        data[f"lag_{lag}"] = data.groupby(["Store","Dept"])["Weekly_Sales"].shift(lag)

    print("Columns after creating features:", data.columns) # Add this line for debugging

    return data
```

```python
# Recursive Multi-step Forecast
# -----------------------------
def recursive_forecast(model, history_df, horizon=12):
    preds = []
    hist = history_df.copy()

    for step in range(horizon):
        X_future = hist[features].iloc[[-1]]
        # Convert to DMatrix before prediction
        dtest = xgb.DMatrix(X_future)
        y_future = model.predict(dtest)[0]

        # Build new row
        next_date = hist["Date"].iloc[-1] + pd.Timedelta(weeks=1)
        new_row = hist.iloc[-1].copy()
        new_row["Date"] = next_date
        new_row["Weekly_Sales"] = y_future

        temp = pd.concat([hist, pd.DataFrame([new_row])])
        temp = create_features(temp).dropna()
        hist = temp.copy()

        preds.append({
            "Store": new_row["Store"],
            "Dept": new_row["Dept"],
            "Date": next_date,
            "Forecast": y_future
        })

    return pd.DataFrame(preds)
```

```python
# âœ… Define feature & target columns (global)
# -----------------------------
# Align with the features list used for training
feature_cols = [
    "Store_enc","Dept_enc","IsHoliday_x","Temperature","Fuel_Price",
    "CPI","Unemployment","weekofyear","month","year","dayofweek",
    "lag_1","lag_2","lag_3","lag_52","roll_mean_4","roll_mean_12"
]
target_col = "Weekly_Sales"

def recursive_forecast(model, history_df, horizon=12):
    """
    Recursive multi-step forecast with updated feature calculation for new rows.

    Args:
        model: trained XGBRegressor or Booster
        history_df: DataFrame containing historical features + target
        horizon: number of steps to forecast ahead

    Returns:
        forecast_df: DataFrame with Date, Forecast
    """
    hist = history_df.copy().sort_values("Date").reset_index(drop=True)

    # Ensure initial history has features calculated
    hist = create_features(hist)
    hist = hist.dropna(subset=feature_cols).reset_index(drop=True) # Drop NaNs from initial feature creation

    if len(hist) == 0:
         raise ValueError("Not enough history after initial feature creation.")

    forecasts = []
    last_known_row = hist.iloc[[-1]].copy()

    # --- Step 3: Recursive prediction ---
    for step in range(horizon):
        # Take the latest row as features
        X_last = last_known_row[feature_cols]

        # Convert to DMatrix before prediction if model is a Booster
        if isinstance(model, xgb.Booster):
             dtest = xgb.DMatrix(X_last)
             y_pred = model.predict(dtest)[0]
        else: # Assume it's an XGBRegressor that can predict on DataFrame
             y_pred = model.predict(X_last)[0]

        # Create next date
        next_date = last_known_row["Date"].iloc[-1] + pd.Timedelta(weeks=1)

        # Append forecast
        forecasts.append({
            "Store": last_known_row["Store"].iloc[-1],
            "Dept": last_known_row["Dept"].iloc[-1],
            "Date": next_date,
            "Forecast": y_pred
        })

        # --- Step 4: Create and append new row with calculated features ---
        new_row_data = last_known_row.iloc[-1].to_dict()
        new_row_data["Date"] = next_date
        new_row_data["Weekly_Sales"] = y_pred # Feed prediction as "truth"

        # Create a temporary dataframe for the new row to calculate its features
        temp_hist = pd.concat([hist, pd.DataFrame([new_row_data])], ignore_index=True)
        # Calculate features only for the newly added row by reapplying create_features
        # and selecting the last row. This recalculates lags/rolling means relative to the *new* end of history.
        temp_hist = create_features(temp_hist)
        last_known_row = temp_hist.iloc[[-1]].copy() # Update last_known_row with the new row and its features
        hist = temp_hist.copy() # Update the history DataFrame as well


    forecast_df = pd.DataFrame(forecasts)
    return forecast_df
```

```python
# -----------------------------
# Multi-step Forecasting for ALL Storeâ€“Dept pairs
# -----------------------------
FORECAST_HORIZON = 12   # weeks ahead

all_forecasts = []
unique_pairs = df[["Store","Dept"]].drop_duplicates()

# Initialize LabelEncoder for 'Type' outside the loop
type_encoder = LabelEncoder()
type_encoder.fit(df['Type'])  # ensure all categories are seen

for _, row in unique_pairs.iterrows():
    store_id, dept_id = row["Store"], row["Dept"]
    last_history = df[(df["Store"]==store_id) & (df["Dept"]==dept_id)].copy()

    if len(last_history) > 60:  # ensure enough history
        try:
            # Apply Label Encoding to 'Type'
            last_history['Type'] = type_encoder.transform(last_history['Type'])

            # Forecast
            forecast_df = recursive_forecast(
                model,
                last_history,
                horizon=FORECAST_HORIZON
            )

            # Add Store and Dept info
            forecast_df["Store"] = store_id
            forecast_df["Dept"] = dept_id

            all_forecasts.append(forecast_df)

        except ValueError as e:
            print(f"Skipping Store {store_id}, Dept {dept_id}: {e}")
        except Exception as e:
            print(f"An error occurred for Store {store_id}, Dept {dept_id}: {e}")

# âœ… Combine all forecasts after loop
all_forecasts_df = pd.concat(all_forecasts).reset_index(drop=True)

print(all_forecasts_df.head())
print(all_forecasts_df.shape)
```
